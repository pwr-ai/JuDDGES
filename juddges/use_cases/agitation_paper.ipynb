```python
from datetime import datetime

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Extract judgment dates and convert to years
judgment_dates = []
for _, row in art_111_df.iterrows():
    try:
        date = row.properties.get("judgment_date")
        if date and isinstance(date, datetime):
            judgment_dates.append(date.year)
    except (AttributeError, TypeError):
        continue

# Create a DataFrame with the years
years_df = pd.DataFrame({"year": judgment_dates})

# Create the histogram
plt.figure(figsize=(12, 6))
sns.histplot(
    data=years_df,
    x="year",
    bins=range(min(judgment_dates), max(judgment_dates) + 2),
    kde=True,
    color="steelblue",
)

plt.title("Distribution of Judgments by Year", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Number of Judgments", fontsize=14)
plt.xticks(range(min(judgment_dates), max(judgment_dates) + 1))
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Add count labels on top of each bar
for p in plt.gca().patches:
    height = p.get_height()
    if height > 0:
        plt.gca().text(
            p.get_x() + p.get_width() / 2.0,
            height + 0.1,
            int(height),
            ha="center",
            fontsize=10,
        )

plt.tight_layout()
plt.show()

# Print some statistics
print(f"Total judgments with dates: {len(judgment_dates)}")
print(f"Earliest judgment: {min(judgment_dates)}")
print(f"Latest judgment: {max(judgment_dates)}")
print(
    f"Most common year: {pd.Series(judgment_dates).value_counts().idxmax()} with {pd.Series(judgment_dates).value_counts().max()} judgments"
)
```

```python
# Get judgments from 2014, 2015
judgments_2014 = art_111_df[
    art_111_df.properties.apply(
        lambda x: x.get("judgment_date")
        and isinstance(x.get("judgment_date"), datetime)
        and x.get("judgment_date").year == 2014
    )
]

judgments_2015 = art_111_df[
    art_111_df.properties.apply(
        lambda x: x.get("judgment_date")
        and isinstance(x.get("judgment_date"), datetime)
        and x.get("judgment_date").year == 2015
    )
]

# Display the number of judgments for each year
print(f"Number of judgments in 2014: {len(judgments_2014)}")
print(f"Number of judgments in 2015: {len(judgments_2015)}")


# Create a function to extract text previews from judgments
def get_text_previews(judgments_df, max_length=500):
    text_previews = []
    for _, row in judgments_df.iterrows():
        if "full_text" in row.properties and row.properties["full_text"]:
            preview = row.properties["full_text"][:max_length]
            text_previews.append(
                {
                    "judgment_id": row.properties.get("judgment_id", ""),
                    "docket_number": row.properties.get("docket_number", ""),
                    "text_preview": preview,
                    "text_length": len(row.properties["full_text"]),
                }
            )
    return pd.DataFrame(text_previews)
```

```python
# Check which judgments contain the word "prezyd" (which can match "prezydent", "prezydencki", etc.)
prezyd_judgments = art_111_df[
    art_111_df.properties.apply(
        lambda x: "full_text" in x
        and isinstance(x["full_text"], str)
        and "prezyd" in x["full_text"].lower()
    )
]

# Display the count of judgments containing "prezyd"
print(f"Number of judgments containing 'prezyd': {len(prezyd_judgments)}")

# Extract basic information about these judgments
prezyd_info = prezyd_judgments.apply(
    lambda row: {
        "judgment_id": row.properties.get("judgment_id", ""),
        "docket_number": row.properties.get("docket_number", ""),
        "court_type": row.properties.get("court_type", ""),
        "judgment_date": row.properties.get("judgment_date", ""),
        "contains_prezyd": True,
        "text_length": len(row.properties.get("full_text", "")),
    },
    axis=1,
)

# Convert to DataFrame for better display
prezyd_df = pd.DataFrame(prezyd_info.tolist())
display(prezyd_df)


# Extract text snippets containing "prezyd" with context
def extract_context(text, keyword, context_size=100):
    """Extract snippets of text containing keyword with surrounding context"""
    if not isinstance(text, str):
        return []

    text_lower = text.lower()
    snippets = []

    start_pos = 0
    while True:
        pos = text_lower.find(keyword, start_pos)
        if pos == -1:
            break

        # Get context around the keyword
        snippet_start = max(0, pos - context_size)
        snippet_end = min(len(text), pos + len(keyword) + context_size)
        snippet = text[snippet_start:snippet_end]

        # Add ellipsis if we're not at the beginning/end of the text
        if snippet_start > 0:
            snippet = "..." + snippet
        if snippet_end < len(text):
            snippet = snippet + "..."

        snippets.append(snippet)
        start_pos = pos + len(keyword)

    return snippets


# Extract and display snippets containing "prezyd" with context
for _, row in prezyd_judgments.iterrows():
    if "full_text" in row.properties and isinstance(row.properties["full_text"], str):
        print(f"\nJudgment ID: {row.properties.get('judgment_id', '')}")
        print(f"Docket Number: {row.properties.get('docket_number', '')}")

        snippets = extract_context(row.properties["full_text"], "prezyd", context_size=100)
        for i, snippet in enumerate(snippets, 1):
            print(f"\nSnippet {i}:")
            print(snippet)
        print("-" * 80)
```
